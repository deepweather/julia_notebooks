```julia
x = -3:0.1:3
```

```julia
f(x) = 3.5 * x ^4 + -2 * x ^ 3 + 2 * x ^ 2 + 3 * x

y = f.(x)
```

```julia
gr()
```

```julia
plot(x,y, label="line")
scatter!(x,y, label="points")
```

```julia
global_temperatures = [14.4, 14.5, 14.8, 15.2, 15.5 , 15.8]
num_pirates =  [45000, 20000, 15000, 5000, 400, 17]
```

```julia
plot(num_pirates, global_temperatures, legend=false)
scatter!(num_pirates, global_temperatures, legend=false)

# reverse x axis
xflip!()

# Add titles and labels

xlabel!("Number of Pirates")
ylabel!("Global Temparature (C)")
title!("Influence of pirate population on the global temperature :D ")
```
